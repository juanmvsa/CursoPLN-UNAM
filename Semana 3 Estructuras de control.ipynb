# 1. Estructuras de control

En todos los lenguajes de programación existen las estructuras de control, las cuales nos permiten definir lógicas dentro de nuestro código. Son indispensables para la automatización.

En las sesiones anteriores teníamos únicamente líneas de código que se ejecutaban secuencialmente (de arriba hacia abajo). Las estructuras de control nos permiten hacer código que se ejecute cierto número de veces o bajo ciertas condiciones dadas, haciendo los programas más poderosos.

# 2.1 if

La secuencia `if` evalua una condición. Si esta es cierta, entonces Python continua con la ejecución del código debajo de dicha secuencia. En cambio, si la condición es falsa, se interrumpe la ejecución del código siguiente, y se sale del bloque de código.

```python
if(2+2)==5:
  print("Esto no se imprimirá")
```

## 2.2 if - else

Al añadir un `else` a un `if`, se agregan condiciones al bloque de código. Si no se ejecuta el `if`, entonces se ejecuta el `else`. Por otra parte, si se ejecuta del `if`, no se ejecuta el `else`.

```python
if(2 + 2) == 5:
  print("Esto no se imprimirá")
else:
  print("Se ejecuta el else")
```

## 2.3 elif

La palabra `elif` pemite añadir cuantas condiciones queramos a nuestro bloque de código. Si uno se ejecuta el primer `if`, entonces se evalúan todas las cláusulas `elif`. Si se cumple alguna, se ejecuta. De lo contrario el bloque ejecuta el `else`.

```python
frutas = ['manzana', 'platano', 'pera', 'uvas']

if ('chile' in frutas):
  print("Hay un chile en la lista de frutas.")
elif ('manzana' in frutas):
  print('Hay una manzana en la lista de frutas.')
elif ('pera' in frutas):
  print('Hay una pera en la lista de frutas.') 
else:
  print("Ninguna condición se cumplió.")
```

# 3. while


Ejecuta el ciclo hasta que la condición se vuelva falsa.

```python
contador = 1

# mientras 1 sea menor a 6, entonces
# 2 < 4 ? sí
while contador <= 6:
  print(contador)
  contador = contador + 1 # contador = 2
```

# 4. continue

Si una condición `if` se cumple, normalmente se ejecuta el ciclo y termina la ejecución. Si se utiliza la palabra `continue`, el ciclo no se interrumpe.

```python
palabra = 'doctor'
len(palabra)
```

```python
vocales = 'aeiou'
palabra = 'doctor'
contador = 0

# itera todas las letras en la palabra
for i in range(len(palabra)):
  # la letra actual es una vocal ? 
  if palabra[i] in vocales:
    continue 
  # aumenta el contador únicamente para no-vocales  
  contador = contador + 1
  # imprime el resultado

  
print('La palabra tiene', contador , 'letras consonantes.')
```

## 5. break

La palabra `break` termina la ejecución de un ciclo, independientemente de que se haya cumplido la condición del `if` inmediatamente anterior.

```python
vocales = 'aeiou'
palabra = 'doctor'
contador = 0

# recorre todas las letras
for i in range(len(palabra)):
  # la letra actual es una vocal ?
  if palabra[i] in vocales:
    break # si se cumple, sal del ciclo
    contador = contador + 1
print('La palabra comienza con', contador, 'letras consonantes.')
```
