# 1. ¿Por qué los lingüistas deben aprender a programar?
La programación es una habilidad muy útil en diversas áreas de la lingüistica y en otras ramas relacionadas, como el estudio del habla, la psicología, psicolingüistica, y estudios cuantitativos en literatura.

*   Constuir todas las combinaciones posibles de palabras compuestas dada una lista de palabras.

```python
import spacy
import es_core_news_sm

nlp = es_core_news_sm.load()
```

```python
texto = ("La farmacéutica estadounidense Pfizer dijo el lunes que empezó ensayos"
  " clínicos en etapa intermedia y tardía de una píldora para prevenir el COVID-19"
  " en personas expuestas al virus. Varias compañías trabajan en posibles"
  " antivirales orales, que imitarían lo que el medicamento Tamiflu hace contra"
  " la influenza y prevendría que la enfermedad avance a un estado severo."
  " Creemos que el combate al virus requerirá tratamientos efectivos para"
  " personas que contraen el virus o han estado expuestos a él, complementando"
  " el impacto que han tenido las vacunas, dijo Mikael Dolsten, jefe de"
  " investigaciones científicas de la compañía.")
```

```python
doc = nlp(texto)

# Analiza la sintaxis
print("Oraciones con sustantivos:", [chunk.text for chunk in doc.noun_chunks])
print("Verbos:", [token.lemma_ for token in doc if token.pos_ == "VERB"])

# Etiquetado
for entity in doc.ents:
    print(entity.text, entity.label_)
```

### Algunas consideraciones sobre cómo nombrar variables:
- Las variables pueden contener todas las letras del alfabeto, todos los números de 0 a 9 y las combinaciones posibles entre estos.
- Por convención, las variables siempre se escriben con minúsculas.
- Se pueden utilizar guiones bajos al nombrar variables.
- Por convención, no se inicia el nombre de una variable con un guión bajo.
- Las variables deben ser nombradas de acuerdo a lo que contengan. Esto para facilitar la lectura del código.

### Ejemplos:
    * variable = 10
    * VARIABLE = 10 (por convención, evitar hacer esto)
    * var1 = 10
    * nombre_apellido = "Juan Vásquez"
    * _nombre = "Juan Vásquez" (evitar nombrar de esta manera)
