Hasta ahora hemos estudiado:

* Lectura y escritura de información en archivos .xsls, .txt y .csv.
* Funciones - Definición y cómo escribir una función para uso particular.

Todos estos elementos permiten construir programas muy complejos usando **objetos**.

Una **clase** es una abstracción computacional que contiene código general. Un objeto es una **implementación** concreta de una clase.

Una clase (y, por lo tanto, un objeto) está conformada por sus **atributos** y **métodos** (o funciones).

Los atributos de una clase son las variables que definen a la clase. Cuando se implementa una clase, el objeto generado adquiere los mismos atributos definidos en la clase.
Los métodos de una clase son las funcionalidades que tiene la clase. Cuando se implementa una clase, el objeto generado adquiere los mismos métodos definidos en la clase.

Estas propiedades permiten tomar código de **librerías**, e implementarlo para nuestras propias necesidades sin tener que volver a escribir todo el código ya hecho dentro de las librerías. Reutilizar código de librerías consiste en implementar objetos a partir de las clases definidas dentro de las librerías.

```python
# Se define primero el nombre de la clase "Perro".
class Perro:
    # Atributos de la clase Perro.
    # Cada atributo se guarda en una variable única.
    raza = ''
    tamaño = ''
    color = ''
    edad = ''
    sonido = '' 

    # Método constructor.
    def __init__(self, raza, tamaño, color, edad, sonido):
        self.raza = raza
        self.tamaño = tamaño
        self.color = color
        self.edad= edad
        self.sonido = sonido

    # Método descriptor.
    def description(self):
        return ("La raza del perro es " + self.raza +
                ", el tamaño del perro es " + self.tamaño +
                ", el color del perro es " + self.color +
                ", la edad del perro es " + self.edad + ".")

    # Método para "ladrar".
    def ladrido(self):
        return ("Este perro emite el sonido " + self.sonido)

    def comer(self):
      if self.raza == 'chow': 
        print("Esta raza come mucho.")
      else:
        print("Raza desconocida.")
```

```python
firulais = Perro('chow', 'grande', 'miel', '7 años', 'ladrido calmado')
```

```python
firulais.description()
```

```python
firulais.ladrido()
```

```python
firulais.comer()
```
