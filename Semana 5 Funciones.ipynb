# 1. Función básica

Las funciones nos permiten escribir código legible y organizado. Cuando un programa empieza a incrementar su tamaño, se vuelve difícil seguirlo. Es por esto que se introduce el concepto de `función` dentro de la programación.

Una `función` (o 'método') es un bloque de código con una funcionalidad específica. Su buen diseño indica que una función debe tener sólo una funcionalidad. Esto permite saber qué hace el bloque de código dentro de la función, sin necesidad de leer todo el bloque de código que constituye a la función.

oop.png

Es recomendable utilizar funciones cuando se sabe que cierto código se va a reutilizar mucho. Esto nos permite simplemente tomar código ya hecho e implementarlo, y no estar escribiendo lo mismo cada vez que se vaya a hacer algo que ya esta programado.

Recordando la primera sesión, Python es un lenguaje muy versátil y usado actualmente, debido a que existen muchas librerías para aplicaciones específicas. Las librerías contienen una gran variedad de funciones. Por lo general, las librerías tienen su documentación, en la cual se explica a detalle cómo implementar las funciones dentro de ella.

En resumen: El concepto de `función` nos permite escribir nuestros propios bloques de código dispuestos a nuestras necesidades particulares. Además nos permite buscar código ya hecho por equipos, e implementar cierto código por medio de las funciones dentro de las librerías.


```python
def mi_funcion():
  print('Esta es una función')
  print("Sólo imprime esto :)")
```

```python
mi_funcion()
```

Para definir una función se debe utilizar la palabra reservada `def`. Luego, un espacio, seguido del nombre que se le quiera dar a la función. Al igual que nombrar variables, nombrar funciones tiene ciertas convenciones. Es preferible nombrar las funciones con palabras que describan su funcionalidad (ej: una función que sume dos números y se llame `suma_numeros`) minúsculas. Se pueden utilizar mayúsculas, minúsculas, números y guiones bajos. Se sugiere iniciar sus nombres con minúsculas (al igual que al nombrar variables).

func.jpeg

Los parámetros de una función son las variables que toma. Puede haber funciones que no tomen ningún parámetro.

Una función puede estar formada por cuantos parámetros se desee. 

La palabra `return` es el lo que el bloque de código dentro de la función genera cuando se acaba su ejecución. 

```python
def suma_numeros(a, b):
  suma = int(a) + int(b)
  return suma
```

```python
print(suma_numeros(5,10))
```

```python
var1 = 50
var2 = 20

suma_nueva = suma_numeros(var1, var2)

print(suma_nueva)
```

# 2. Función con estructuras de control

```python
# Función que toma un número entre 1 y 10.
# Si el número no esta en ese rango se imprime un mensaje de error.
def numero():
  num = int(input('Introduce un número del 1 al 10: '))
  if num < 1:
    print("El número introducido es menor a 1. Error.")
  elif num > 10:
    print("El número introducido es mayor. Error.")
  
  return num
```

```python
numero()
```

```python
def numero_con_argumento(a):
  if a < 1:
    print("El número introducido es menor a 1. Error.")
  elif a > 10:
    print("El número introducido es mayor. Error.")
  
  return a
```

```python
numero_con_argumento(100)
```

```python
# Función que pega dos oraciones.
def pega_oraciones(oracion1, oracion2):
  return oracion1 + oracion2
```

```python

pega_oraciones("Mi nombre es", " Juan")
```

```python
# Función que separa una oración
# cuando encuentra un punto.
def separa_oracion(oracion): 
  separado = oracion.split('.') 
  return separado
```

```python
separa_oracion("Esta.Función. Separa. Cuando encuentra. un punto.")
```

```python
# Función que cuenta el número de 'a's en 
# un texto ingresado manualmente por el usuario
def cuenta(texto):
  contador = 0
  for letra in texto:
    if letra == 'a':
      contador = contador + 1
  return contador
```

```python
texto = """ El Procesamiento del Lenguaje Natural es el campo de conocimiento 
de la Inteligencia Artificial que se ocupa de la investigar la manera de 
comunicar las máquinas con las personas mediante el uso de lenguas naturales, 
como el español, el inglés o el chino."""

cuenta(texto)
```
